```python
import sklearn
import pandas as pd
import sqlite3
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from datetime import datetime
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.preprocessing import StandardScaler
from tensorflow import keras
import tensorflow as tf
import pickle
import os
import joblib

import warnings
warnings.filterwarnings('ignore')

MODEL_PATH = 'TFAuto'
model = keras.models.load_model(MODEL_PATH)
scaler = joblib.load(MODEL_PATH + '/TFScaler.pkl')
```

```python
conn = sqlite3.connect('../db.sqlite3')
```

```python
df_raw = pd.read_sql_query("SELECT * FROM scraper_auto", conn)
```

```python
df = df_raw[['bouwjaar', 'kilometer_stand', 'vermogen', 'is_handgeschakeld', 'is_benzine']]
df
```

```python
df_scaled = scaler.transform(df)
```

```python
df_raw['Prediction'] = model.predict(df_scaled)
df_raw['Score'] = df_raw['Prediction'] - df_raw['prijs']

df_raw.sort_values(by='Score', ascending=False).dropna().head(20)
```
